```python
import random
def choose_oddoreven():
    ply1=input("CHOOSE ODD OR EVEN : ").lower()
    com1=random.choice(["odd","even"])
    if ply1=="odd":
        com1="even"
    else:
        com1="odd"
    return ply1,com1
def throw_fingers():
    f_ply1=int(input("Enter a number between 1-6 :"))
    f_com1=random.randint(1,6)
    return f_ply1,f_com1
def check_odd_even(sum_count):
    if sum_count % 2 == 0:
        return 'even'
    else:
        return 'odd'
def choose_batorbowl(ply1, com1, f_ply1, f_com1):
    sum_count = f_ply1 + f_com1
    if check_odd_even(sum_count) == ply1:
        print("You won the Toss >> choose whether to BAT or BOWL")
        ply1_role = input("BAT OR BOWL: ").lower()
        if ply1_role == "bat":
            com1_role = "bowl"
        else:
            com1_role = "bat"
    else:
        print("Computer won the toss")
        com1_role = "bat" if com1 == 'even' else "bowl"
        if com1_role == "bat":
            print("Computer choose to bat first")
        else:
            print("Computer choose to bowl first")
        ply1_role = "bowl" if com1_role == "bat" else "bat"
    return ply1_role, com1_role
def batting(ply1_role):
    if ply1_role == "bat":
        bat()
    else:
        bowl()
def bat():
    runs = 0
    botruns = 0
    gamerunning = False
    while not gamerunning:
        player = int(input("It's your batting. Choose a number between 1-6: "))
        bot = random.randint(1, 6)
        print("Bot is bowling", bot)
        runs += player
        print("Current run is", runs)

        if player == bot:
            print("User is out")
            while True:
                player = int(input("You may do bowling: "))
                bot = random.randint(1, 6)
                print("Bot batting:", bot)
                botruns += bot
                print("Current runs of bot are:", botruns)

                if botruns > runs:
                    print("Bot won the match. Better luck next time")
                    gamerunning = True
                    break
                elif player == bot:
                    print("Successfully taking the wicket of AI")
                    print("You have won the match by", (runs - botruns), "runs")
                    gamerunning = True
                    break
            break  
def bowl():
    runs = 0
    botruns = 0
    gamerunning = False
    while not gamerunning:
        bot = random.randint(1, 6)
        player = int(input("It's your bowling. Choose a number between 1-6: "))
        print("Bot is batting", bot)
        botruns += bot
        print("Current runs of bot are:", botruns)

        if player == bot:
            print("Bot is out")
            while True:
                player = int(input("You may do batting: "))
                bot = random.randint(1, 6)
                print("Bot batting:", bot)
                runs += player
                print("Current run is", runs)

                if runs > botruns:
                    print("You won the match. Congratulations!")
                    gamerunning = True
                    break
                elif player == bot:
                    print("Successfully taking the wicket of bot")
                    print("Bot has won the match by", (botruns - runs), "runs")
                    gamerunning = True
                    break
            break  
def bot_batting(com1_role):
    if com1_role == "bat":
        bowl()
    else:
        bat()
def replay():
    return input('Do you want to play again? Enter Yes or No: ').lower().startswith('y')
while True:
    ply1, com1 = choose_oddoreven()
    f_ply1, f_com1 = throw_fingers()
    sum_count = f_ply1 + f_com1
    check_odd_even(sum_count)
    ply1_role, com1_role = choose_batorbowl(ply1, com1, f_ply1, f_com1)
    if ply1_role=="bat"or"bowl":
        batting(ply1_role)
        gamerunning=False
    elif com1_role=="bat"or"bowl":
        bot_batting(com1_role)
        gamerunning=False
        break
    if not replay():
        break
```
